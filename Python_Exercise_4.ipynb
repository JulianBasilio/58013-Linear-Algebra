```python
###Vector Operations and It's applications
import numpy as np
import matplotlib.pyplot as plt
### Initialize given arrays
A= np.array([
    [-0.4],
    [4.3],
    [-0.6]
])
B= np.array([
    [-0.2, 0.2, 1]
])

C= np.array([
    [-0.2, 2.1, -1.5]
])
print("Array A is",A)
print("Array B is",B)
print("Array C is",C)
print("The shape of Array A is",np.shape(A))
print("The shape of Array B is",np.shape(B))
print("The shape of Array C is",np.shape(C))
```

```python
A= A.reshape(1,3) ##Reshapes Array A from a (3,1) array to a (1,3) array
Add= np.add(A,B,C) ### Addition of Vectors
print("The addition of three vectors is",Add)
```

```python
A= A.reshape(1,3) ##Reshapes Array A from a (3,1) array to a (1,3) array
Sub= np.subtract(A,B,C) ### Subtraction of Vetors
print("The subtraction of three vectors is",Sub)
```

```python
A= A.reshape(1,3) ##Reshapes Array A from a (3,1) array to a (1,3) array
Times= np.multiply(A,B,C) ### Multiplication of Vectors
print("The multiplication of three vectors is",Times)
```

```python
A= A.reshape(1,3) ##Reshapes Array A from a (3,1) array to a (1,3) array
Div= np.divide(A,B,C) ### Division of Vetors
print("The division of three vectors is",Div)
```

```python
### Graph
fig= plt.figure()
ax= fig.add_subplot(111, projection= "3d")
a= [-0.6,  4.5 , 0.4] ###Result of A+B+C
s= [-0.2,  4.1 , -1.6]##Result of A-B-C
m= [ 0.08,  0.86, -0.6 ]##Result of A*B*C
d= [ 2.,  21.5, -0.6] ###Result of A/B/C
ax.quiver(0,0,0,a[0], a[1], a[2], color="r", arrow_length_ratio=0.1) ### Addition vector =red
ax.quiver(0,0,0,s[0], s[1], s[2], color="g", arrow_length_ratio=0.1) ### Subtraction vector = green
ax.quiver(0,0,0,m[0], m[1], m[2], color="b", arrow_length_ratio=0.1) ### Multiplication vector = blue
ax.quiver(0,0,0,d[0], d[1], d[2], color="y", arrow_length_ratio=0.1) ### Division Vector = yellow


ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Graph of Vectors")
plt.show()
```
